# RFM Analysis ( Customer Segmentation ) of a online Store's Retail Data

```python
# Importing necessary libraries for data manipulation, visualization, and machine learning:
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading the retail dataset from a CSV file for analysis
df = pd.read_csv('OnlineRetail.csv', encoding = 'unicode_escape')
df.head()
```

```python
# Checking the data types of each column to understand the structure of the dataset
df.dtypes
```

## Prepare the data for RFM (Recency, Frequency and Monetary) Analysis with K-Mean Clustering Algorithm

```python
#Prepare the data for Recency:
df['InvoiceDate'].tail(1)
```

```python
currentDate = pd.to_datetime("2012-01-01")
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df
```

```python
currentDate
```

```python
df.dtypes
```

```python
# Finding the last purchased date of the customers

maxdate = df.groupby(['CustomerID']).max()[['InvoiceDate']]
maxdate
```

```python
# Finding the Destance from Current Date of Last purchased date

(currentDate - maxdate['InvoiceDate']).dt.days
```

```python
maxdate['InvoiceAge'] = (currentDate - maxdate['InvoiceDate']).dt.days
maxdate
```

```python
#Finding the Recenccy of each Customer

recency = maxdate.drop('InvoiceDate', axis=1)
recency
```

```python
#Prepare the data for Frequency:

freq = df.drop_duplicates(subset="InvoiceNo")
freq
```

```python
#Finding the Frequency of each Customer

freq = freq.groupby(["CustomerID"]).count()[["InvoiceNo"]]
freq
```

```python
#Prepare the data for Monetary:

#Add new column 'Total' with the multiplication of Quantity & UnitPrice
df["Total"]=df["Quantity"]*df["UnitPrice"]
df
```

```python
# Finding the Monetary of each Customer

money=df.groupby(["CustomerID"])[["Total"]].sum()
money
```

```python
#Creating the RFM Table

RFM = pd.concat([recency,freq,money], axis=1)
RFM
```

```python
#Rename the column to Recency, Frequency and Monetary

RFM.columns = ['Recency', 'Frequency', 'Monetary']
RFM
```

## Creating the model for RFM Analysis ( Customer Segmentation )

```python
#Scaling the data
# Standardizing the RFM values before clustering

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_RFM=scaler.fit_transform(RFM)
scaled_RFM
```

```python
scaled_RFM.shape
```

```python
from sklearn.cluster import KMeans
```

# Finding the value of Optimal K

```python
k_range = range(2,11)
sse = []
for k in k_range:
    km = KMeans(n_clusters=k) 
    km.fit(scaled_RFM)
# Using the Elbow Method to determine optimal number of clusters
    sse.append(km.inertia_)
```

```python
plt.plot(k_range, sse, 'r-*')
plt.xlabel ='k'
plt.ylabel = 'Sum of Squared Error'
```

# Object Creation and Data fitting to the algorithm

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(scaled_RFM)
```

```python
RFM["Clusters"]=(kmeans.labels_)
RFM
```

```python
fig = plt.figure(figsize=(10,8))
ax = plt.subplot(111, projection='3d', label="bla")
ax.scatter(RFM["Recency"],RFM["Frequency"],RFM["Monetary"], s=40, c=RFM["Clusters"], marker='o', cmap = "Accent")
ax.set_title("The Plot Of The Clusters")
# Displaying the plot
plt.show()
```

```python
RFM_mean=RFM.groupby("Clusters").mean()[["Recency","Frequency","Monetary"]]
RFM_mean
```

```python
def Customer_Segmentation(RFM_Cluster):
    if RFM_Cluster["Clusters"]==2:
        return 'Diamond'
    elif RFM_Cluster["Clusters"]==1:
        return 'Gold'
    else:
        return 'Silver'
```

```python
RFM['group']=RFM.apply(Customer_Segmentation, axis=1)
RFM
```

```python
result = pd.DataFrame(RFM.group.value_counts())
result = result.reset_index()
result
```

```python
# Visualizing data with Seaborn plots
datalabel=sns.barplot(x="group", y="count", data=result, estimator=np.sum)
for i in datalabel.containers:
    datalabel.bar_label(i,)
```
